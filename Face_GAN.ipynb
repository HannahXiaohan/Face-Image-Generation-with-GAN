```python
import os
from collections import OrderedDict
from options.test_options import TestOptions
from data.data_loader import CreateDataLoader
from models.models import create_model
import util.util as util
from util.visualizer import Visualizer

opt = TestOptions().parse(save=False)
opt.display_id = 0 # do not launch visdom
opt.nThreads = 1   # test code only supports nThreads = 1
opt.batchSize = 1  # test code only supports batchSize = 1
opt.serial_batches = True  # no shuffle
opt.no_flip = True  # no flip
opt.in_the_wild = True # This triggers preprocessing of in the wild images in the dataloader
opt.traverse = True # This tells the model to traverse the latent space between anchor classes
```

```python
data_loader = CreateDataLoader(opt)
dataset = data_loader.load_data()
visualizer = Visualizer(opt)
```

```python
opt.name = 'females_model' # change to 'females_model' if you're trying the code on a female image
model = create_model(opt)
model.eval()
```

```python
import torch
import torch.nn.utils.prune as prune

def safe_prune(module, amount=0.3):
    """
    Try pruning a module safely. Supports EqualConv2d, EqualLinear, Conv2d, Linear.
    """
    # Try pruning conv inside custom modules (EqualConv2d)
    if hasattr(module, 'conv') and isinstance(module.conv, torch.nn.Conv2d):
        prune.l1_unstructured(module.conv, name='weight', amount=amount)
        print(f"✅ Pruned {type(module).__name__}.conv")

    # Try pruning linear inside custom modules (EqualLinear)
    elif hasattr(module, 'linear') and isinstance(module.linear, torch.nn.Linear):
        prune.l1_unstructured(module.linear, name='weight', amount=amount)
        print(f"✅ Pruned {type(module).__name__}.linear")

    # Direct Conv2d or Linear modules
    elif isinstance(module, (torch.nn.Conv2d, torch.nn.Linear)):
        prune.l1_unstructured(module, name='weight', amount=amount)
        print(f"✅ Pruned {type(module).__name__}")

def apply_pruning_fully(model, amount=0.3):
    """
    Apply safe pruning to the entire model.
    """
    for name, module in model.named_modules():
        try:
            safe_prune(module, amount)
        except Exception as e:
            print(f"⚠️ Skipped {name} ({type(module).__name__}): {e}")

# 👇 Now apply to model.netG
apply_pruning_fully(model.netG, amount=0.3)
torch.save(model.netG.state_dict(), 'model_pruned.pth')
print("Pruned model size:", os.path.getsize('model_pruned.pth')/1024/1024, "MB")
print("✅ Full safe pruning done on Generator!")
```
